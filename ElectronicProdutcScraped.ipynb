```python
class Electronic:
    Name = ""
    Price = ""
    Ratings = ""
    Reviews = 0
    Discount = 0
    Country = ""
    Brand = ""
    Installments = ""
    def __init__(self , name , price , reviews , ratings , discount , brand , country , installments):
        self.Name = name
        self.Price = price
        self.Ratings = ratings
        self.Reviews = reviews
        self.Discount = discount
        self.Brand = brand
        self.Country = country
        self.Installments = installments
        
array_list=[]
driver = webdriver.Chrome(executable_path='D:\Semester 3\DSA LAB\week 2\Chrome Driver\chromedriver.exe')

                
def smartPhonescrap():
    
    smartPhonepages = list(range(1 , 79))
    for smP in smartPhonepages:
        driver.get("https://www.daraz.pk/smartphones/?page={}&spm=a2a0e.searchlistcategory.cate_1.1.2ff05e4eQPmEY5".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = ""
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i > 12:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                    i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
                
                
                
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text

            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)
            
    header = ['Name', 'Price'  , 'Ratings' , 'Reviews' , 'Discount' , 'Brand' , 'Country' , 'Installments']
    with open("PrdEln.csv", "w", newline='') as f:
        writer = csv.writer(f, delimiter=',')
        writer.writerow(header)

    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)

array_list=[] 
def featurePhonescrap():
    
    featurePhonepages = list(range(1 , 29))
    for smP in featurePhonepages:
        driver.get("https://www.daraz.pk/featurephones/?page={}&spm=a2a0e.searchlistcategory.cate_1.2.5396c414KTrLcx".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            
            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
                
                
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)
            


    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)
featurePhonescrap()  
array_list=[]
def landLinePhonescrap():    
    landLinePhonepages = list(range(1 , 54))
    for smP in landLinePhonepages:
        driver.get("https://www.daraz.pk/corded-phones/?page={}&spm=a2a0e.searchlistcategory.cate_1.4.b3af7242YlNkvy".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
                
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text

            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)
            

    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)
landLinePhonescrap()
array_list=[]       
def laptopscrap():
    laptoppages = list(range(1 , 23))
    for smP in laptoppages:
        driver.get("https://www.daraz.pk/laptops/?page={}&spm=a2a0e.searchlistcategory.cate_1.5".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
                
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)


    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)
laptopscrap()          

array_list=[]
def desktopscrap():  
    desktoppages = list(range(1 , 9))
    for smP in desktoppages:
        driver.get("https://www.daraz.pk/desktop-computer/?page={}&spm=a2a0e.searchlistca".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            
            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
                
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text

            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)


    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)
            
desktopscrap()
array_list=[]
def smartWatchscrap():    
    smartWatchpages = list(range(1 , 103))
    for smP in smartWatchpages:
        driver.get("https://www.daraz.pk/smart-watches/?page={}&spm=a2a0e.searchlistcategory.cate_1.7.b4914c06WQnbnH".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            
            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)


    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)

            
smartWatchscrap()
array_list=[]
            
def gamingConsolescrap():
    gamingConsolepages = list(range(1 , 103))
    for smP in gamingConsolepages:
        array_list=[]
        driver.get("https://www.daraz.pk/gaming-consoles/?page={}&spm=a2a0e.searchlistcategory.cate_1.8.6aee6399VZLtuW".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text

            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)


        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
            
gamingConsolescrap()
array_list=[]

def cameracrap():    
    camerapages = list(range(1 , 103))
    for smP in camerapages:
        array_list=[]
        driver.get("https://www.daraz.pk/camera/?page={}&spm=a2a0e.searchlistcategory.cate_1.9.21065d00etwuh1".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
            
cameracrap()
array_list=[]          
def mobileAssessrycrap():   
    mobileAssessrypages = list(range(1 , 103))
    for smP in mobileAssessrypages:
        driver.get("https://www.daraz.pk/mobiles-tablets-accessories/?page={}&spm=a2a0e.searchlistcategory.cate_2.1.1f035df8TLn9Gd".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)

mobileAssessrycrap()
array_list=[]
            
def phoneCamerascrap():    
    phoneCamerapages = list(range(1 , 103))
    for smP in phoneCamerapages:
        driver.get("https://www.daraz.pk/phone-camera-flash-lights/?page={}&spm=a2a0e.searchlistcategory.cate_2_1.10.439c365dpxnsSX".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)
            
phoneCamerascrap()
array_list=[]
def mobileCoverscrap():    
    mobileCOverspages = list(range(1 , 103))
    for smP in mobileCOverspages:
        driver.get("https://www.daraz.pk/mobile-cases-covers/?page={}&spm=a2a0e.searchlistcategory.cate_2_1.1.137d5a2f9lThVf".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)


    for i in array_list:
        data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
        with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=',')
            writer.writerow(data)
mobileCoverscrap()
array_list=[]

def carChargerscrap():    
    carChargerpages = list(range(1 , 103))
    for smP in carChargerpages :
        array_list=[]
        driver.get("https://www.daraz.pk/car-chargers/?page={}&spm=a2a0e.searchlistcategory.cate_2_1.8.31c615c6wDiQxh".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)

            
carChargerscrap()
array_list=[]

def otherComputerAssessryscrap():    
    otherComputerAssessrypages = list(range(1 , 82))
    for smP in otherComputerAssessrypages:
        array_list=[]
        driver.get("https://www.daraz.pk/other-computer-accessories/?page={}&spm=a2a0e.searchlistcategory.cate_2_5.1.372915c6ByRzzG".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            
            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)


        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)


            
            
otherComputerAssessryscrap()         
array_list=[]

def pcAudioscrap():    
    pcAudiopages = list(range(1 , 103))
    for smP in pcAudiopages:
        array_list=[]
        driver.get("https://www.daraz.pk/pc-audio/?page={}&spm=a2a0e.searchlistcategory.cate_2_5.5.680f79e8TK6eLN".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
    
pcAudioscrap()
array_list=[]
def ledTvscrap():
    ledTvpages = list(range(1 , 9))
    for smP in ledTvpages:
        array_list=[]
        driver.get("https://www.daraz.pk/led-tvs/?page={}&spm=a2a0e.searchlistcategory.cate_3.1.f4bc5e2a3PnmUo".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
    

ledTvscrap()        
array_list=[]  
def tvAssessryscrap():    
    tvAssessrypages = list(range(1 , 103))
    for smP in tvAssessrypages:
        array_list=[]
        driver.get("https://www.daraz.pk/tv-accessories/?page={}&spm=a2a0e.searchlistcategory.cate_3.4.dff23805VNJrCV".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
    
            
tvAssessryscrap()
array_list=[]
def ironAppliancescrap():
    ironpages = list(range(1 , 103))
    for smP in ironpages:
        array_list=[]
        driver.get("https://www.daraz.pk/ironing-laundry-appliances/?page={}&spm=a2a0e.searchlistcategory.cate_3_9.1.1f977a40OBxZ2U".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
    

            
ironAppliancescrap()
array_list=[]
def vacumCleanerscrap():
    vacumCleanerpages = list(range(1 , 103))
    for smP in vacumCleanerpages:
        array_list=[]
        driver.get("https://www.daraz.pk/vacuum-cleaners/?page={}&spm=a2a0e.searchlistcategory.cate_3_10.1.4534682dZOx2JZ".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 8:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)


vacumCleanerscrap()       
array_list=[]
def washerDryerscrap():
    washerDryerpages = list(range(1 , 50))
    for smP in washerDryerpages:
        array_list=[]
        driver.get("https://www.daraz.pk/washers-dryers/?page={}&spm=a2a0e.searchlistcategory.cate_3_12.3.155773f24EVwsd".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")
            

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)

washerDryerscrap()       
array_list=[]           
def mensWatchscrap():
    mensWatchpages = list(range(1 , 67))
    for smP in mensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/mens-chronograph/?page={}&service=OS&spm=a2a0e.searchlistcategory.cate_10_1.2.78fa21c2Sgb2Ap".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = ""
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text
            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
        

mensWatchscrap()
array_list=[]
def womensWatchscrap():
    womensWatchpages = list(range(1 , 50))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/womens-fashion-watches-wsj/?page={}&service=OS&spm=a2a0e.searchlistcategory.cate_10_2.2.717221c204DeL3".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
        
womensWatchscrap()

array_list=[]
def securityCamerascrap():
    womensWatchpages = list(range(1 , 103))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/security-cameras/?page={}&spm=a2a0e.searchlistcategory.cate_1.10.691c6399HevvKV".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
                
securityCamerascrap()

array_list=[]
def freezerscrap():
    womensWatchpages = list(range(1 , 6))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/freezers/?page={}&spm=a2a0e.searchlistcategory.cate_3_6.4.64e23382hZ4v9e".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)

freezerscrap()
array_list=[]
def refregratorscrap():
    womensWatchpages = list(range(1 , 39))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/refrigerators/?page={}&spm=a2a0e.searchlistcategory.cate_3_6.1.12547cd9inuYyG".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)

refregratorscrap()


def ovenscrap():
    womensWatchpages = list(range(1 , 9))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/large-ovens/?page={}&spm=a2a0e.searchlistcategory.cate_3_6.3.3f2d3224Ztmg5g".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)

ovenscrap()


def airConditionerscrap():
    womensWatchpages = list(range(1 , 25))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/air-conditioners/?page={}&spm=a2a0e.searchlistcategory.cate_3_8.2.49972140uBNYSD".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)

airConditionerscrap()

def aircoolerscrap():
    womensWatchpages = list(range(1 , 86))
    for smP in womensWatchpages:
        array_list=[]
        driver.get("https://www.daraz.pk/portable-ac/?page={}&spm=a2a0e.searchlistcategory.cate_3_8.3.4fad1547Yvwc08".format(smP))
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'c5TXIP'}):
            brand = " "
            spaceFound = 0
            i = 0
            name=a.find('div' ,attrs={'class':'c16H9d'})
            while spaceFound == 0:
                if name.text[i] == " " or i >= 10:
                    spaceFound = spaceFound+1
                else:
                    brand = brand+name.text[i]
                i = i+1
            price0 = a.find('span' ,attrs={'class':'c13VH6'})
            price = price0.text.replace("Rs. ","")

            reviews = a.find('span' ,attrs={'class':'c3XbGJ'})
            if reviews==None:
                reviews="0"
            else:
                reviews = reviews.text.replace("-","")

            discount = a.find('span' ,attrs={'class':'c1hkC1'})
            if discount==None:
                discount="  0%"
            else:
                discount= discount.text
            country = a.find('span' ,attrs={'class':'c2i43- '})
            if country == None:
                country = "Pakistan"
            else: 
                country = country.text

            ratings = a.find('ul' ,attrs={'class':'ant-rate ant-rate-disabled'})
            if ratings==None:
                ratings="0"
            else:
                ratings = ratings.text
                
            installments = a.find('div' ,attrs={'class':'c3ubLI'})
            if installments == None:
                installments = "No available"
            else: 
                installments = installments.text
            obj = Electronic(name.text , price , ratings , reviews , discount , brand , country , installments)
            array_list.append(obj)

        for i in array_list:
            data=[i.Name,i.Price, i.Ratings , i.Reviews  , i.Discount , i.Brand , i.Country , i.Installments]
            with open("PrdEln.csv", "a", newline='',encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=',')
                writer.writerow(data)
aircoolerscrap()
```
